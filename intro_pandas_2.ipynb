```python
# Show the first five records (rows) in the dataframe 'df2'
df2.head()
```

```python
# Create a new column named 'Gross Profit' and let it be the difference between revenue and cost of sales
df2['gross_profit'] = df2['revenue'].values - df2['cost_of_sales'].values
df2.head()
```

```python
# You can add a column with a scalar value and it will add the scalar to all rows in the column
df2['gross_profit'] + 4
```

```python
# Use the df.apply method to create a column. The df.apply method accepts either a user-defined function 
# or a method from an existing module such as numpy, math, etc
import numpy as np

df2['sqrt_gp'] = df2['gross_profit'].apply(np.sqrt)
df2.head()
```

```python
def power_two(x):
    return x ** 2

df2['power_two'] = df2['gross_profit'].apply(power_two)
df2.head()
```

```python
# You can use Lambda functions whenever a method requires a function. The syntax of a lambda function is:
# lambda parameter: return value
def power_four(x):
    return x ** 4 

df2['power_four'] = df2['gross_profit'].apply(lambda x: x**4)
df2.head()
```

```python
# You can use the df.applymap function to apply a function to all the columns in a dataframe. Let us create 
# a sample dataframe named 'cdf' and use the df.applymap method to get the fourth power of the values in the column
cdf = pd.DataFrame({'A': [1,2,3,4], 'B': [5,6,7,8]})
cdf
```

```python
cdf.applymap(func=lambda x: x**4,na_action='ignore')
```

```python
# You can read a text file from your local machine and make it into a dataframe. Unlike CSV and Excel files that have
# natural tables/rows & columns, if you need to create a dataframe from text, you will need to use Python's Text Stream
# function called 'open'.
oge_txt = open(r"C:\Users\ugo\Desktop\sql_csv_files\oge.txt")
output = oge_txt.readline()
```

```python
# Use the output from the readline() method as the row of the dataframe.
oge_df = pd.DataFrame({'column_1': [output]})
oge_df
```
